# Visualising Stock data using Pandas and Resampling

Ever wondered when we get the stock data of a particular Stock in a time frame.
We always have to work on codes using function like Groupby to form quarterly or monthly data.
But like most of the methods, there is a simpler way of doing things.
And that is Resampling function in pandas.
It allows a convenient method for frequency conversion and resampling of time series.
More details can be found on the following link:

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.resample.html

```python
# let us import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import datetime
%matplotlib inline
```

```python
# Import the data from yahoo finance of Apple stocks for a period of 10 years:
df = web.DataReader('AAPL', data_source='yahoo' , start='2009-01-01')
```

```python
df.head()
```

Resampling is a easy technique to visaulaise the stock data

# Time resampling
This allows to create dataframes with specified requirements:
"Q" -> Returns quarterly data
"A" -> Returns Yearly data and etc.

```python
df.resample(rule='A')
```

```python
# returns yearly data:

df.resample(rule='A').mean()
```

```python
# returns quarterly data:

df.resample(rule='Q').mean()
```

```python
# returns Business quarter endfrequency data:

df.resample(rule='BQ').mean()
```

```python
df.resample(rule='A').max()
```

```python
# creating our own custom function to get the data of each years first day trading:

def first_day(entry):
    return entry[0]

df.resample(rule='A').apply(first_day)
```

```python
# Visualising:
```

```python
df['Close'].resample('A').mean().plot(kind='bar', figsize=(15,8), color='r')
```

```python
df['Close'].resample('M').mean().plot(kind='bar', figsize=(18,8))
```

```python
df['Close'].resample('Q').mean().plot(kind='bar', figsize=(18,8))
```
